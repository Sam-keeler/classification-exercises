```python
# Is the sepal length significantly different in virginica compared to versicolor? 
# Run a statistical experiment to test this.
train
```

```python
versicolor = train[train.species == 'versicolor']
```

```python
virginica = train[train.species == 'virginica']
```

```python
a = .05
```

```python
t, p = stats.ttest_ind(virginica.sepal_length, versicolor.sepal_length)
```

```python
# Null Hypothesis = There is no difference in the sepal length between Virginica and Versicolor
# Alternative hypothesis = There is a difference in sepal length between Virginica and Versicolor
if p < a:
    print("We reject the null hypothesis there is no difference in the sepal length between Virginica and Versicolor")
    print("We move forward w/ the alternative hypothesis")
else:
    print("We fail to reject the null hypothesis")
```
