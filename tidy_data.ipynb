```python
# Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations
# of different recipes, oven rack positions, and oven temperatures.

cake = pd.read_csv('untidy-data/cake_recipes.csv')
cake
```

```python
# Tidy the data as necessary.

cake['id2'] = cake['recipe:position']
```

```python
cake.set_index('id2',inplace=True)
cakeT = cake.T
cake.T
```

```python
cakeT.drop(index = 'recipe:position')
```

```python
cakeT.columns = cakeT.iloc[0]
```

```python
cakeT.drop(index = 'recipe:position', inplace = True)
```

```python
cakeT.rename(columns = {'a:bottom': 'A_bottom', 'b:bottom': 'B_bottom', 'c:bottom': 'C_bottom',
                        'd:bottom': 'D_bottom', 'a:top': 'A_top', 'b:top': 'B_top', 'c:top': 'C_top',
                       'd:top': 'D_top'}, inplace = True)
```

```python
cakeT
```

```python
df2 = cakeT[['A_bottom', 'B_bottom', 'C_bottom', 'D_bottom']]
df1 = cakeT[['A_top', 'B_top', 'C_top', 'D_top']]

df1.columns = ['A','B', 'C', 'D']
df2.columns = ['A','B', 'C', 'D']

```

```python
df1 = df1.rename({'225': '225(Top)', '250': '250(Top)', '275': '275(Top)', '300': '300(Top)'}, axis='index')
```

```python
df2 = df2.rename({'225': '225(Bottom)', '250': '250(Bottom)', '275': '275(Bottom)', '300': '300(Bottom)'}, axis='index')
```

```python
df1
```

```python
cakeTa = df1.append(df2)
```

```python
cakeTa
```

```python
# Which recipe, on average, is the best?
# B

cakeTa.mean()
```

```python
# Which oven temperature, on average, produces the best results? 
# 275

cakeT.mean(axis = 1)
```
